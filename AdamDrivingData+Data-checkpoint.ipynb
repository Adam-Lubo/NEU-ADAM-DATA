```python
myheader = ["time", "Level", "Speed", "Z position", "X position", "Y Rotation", "Steering", "Acc.", "Brake", "Event ID", "n/a"]
df = pd.read_csv("07292022_0226_data.txt", skiprows = 0, header=0, names = myheader)
print("Data: Data Frame")
print(df.head(32))
print(df.describe())
```

```python
startts = datetime(2022, 7, 29, 2, 26, 51)

datestr = "07/29/2022 02:26:51.278"
d = datetime.strptime(datestr, "%m/%d/%Y %H:%M:%S.%f")

#print(datetime.strptime("%M/%d/%y df["time"] - startts)
df['time'] = pd.to_datetime(df['time'], format="%m/%d/%Y %H:%M:%S.%f")
df['offset'] =  (df['time'] - startts)
df = df.set_index("offset")
df.head()
```

```python
df = df.set_index('offset').cumsum(axis=1)
```

```python
#\d- regular expresson match digit
def get_question(row):
    pattern = re.compile("Question_(\d)Select_(\d)")
    m = pattern.match(row["Event ID"])
    if m:
        return m[1]
    else:
        return None
   
def get_answer(row):
    pattern = re.compile("Question_(\d)Select_(\d)")
    m = pattern.match(row["Event ID"])
    if m:
        return m[2]
    else:
        return None

#when first question
def firstquestion(row):
    if row["Speed"] == 0 and row["Brake"] == 0:
        return 1
    else:
        return 0

   
df = df.assign(question=df.apply(get_question, axis=1))
df = df.assign(answer=df.apply(get_answer, axis=1))
df = df.assign(firstquestion=df.apply(firstquestion, axis=1))
df.iat[0, 13] = 0
df.head()
```

```python
found_zero = 0
starting_timestamps = []
ending_timestamps = []
prev_row = 0

for index, row in df.iterrows():
    if row["firstquestion"] == 1 and found_zero == 0:
        found_zero = 1
        starting_timestamps.append(index)
    elif row["firstquestion"] == 0 and found_zero ==1:
        found_zero = 0
        ending_timestamps.append(prev_row)
    prev_row = index

if len(starting_timestamps) != len(ending_timestamps):
    print("Lists are dif")

print(starting_timestamps)

for index, ts in enumerate(starting_timestamps):
    print(df["time"].loc[ts], " to ", df["time"].loc[ending_timestamps[index]])
```

```python
driving_segs = {}
driving_segs["1-2"] = { "tdf": [], "ts" : [] }
driving_segs["3"] = { "tdf": [], "ts" : [] }
driving_segs["4-5"] = { "tdf": [], "ts" : [] }  



for index, ts in enumerate(starting_timestamps):
    tdf = df[starting_timestamps[index]:ending_timestamps[index]]
    answer = set ( tdf["answer"][tdf["question"] == "1"].to_list()).pop()
    print ( "Starting time", ts, " answer is ", answer)
    
    pdf = df[ts - timedelta(seconds=20): ts]  
    print (pdf.describe())
    print ("=" * 30)
    
    key = ""
    if answer == "1" or answer == "2":
        key = "1-2"
    elif answer == "3":
        key = "3"
    elif answer == "4" or answer == "5":
        key = "4-5"
    else:
        print ("funny value?", answer)
        
    driving_segs[key]["tdf"].append(pdf)
    driving_segs[key]["ts"].append(ts)

driving_segs["1-2"]["df"] = pd.concat(driving_segs["1-2"]["tdf"])
driving_segs["3"]["df"] = pd.concat(driving_segs["3"]["tdf"])
driving_segs["4-5"]["df"] = pd.concat(driving_segs["4-5"]["tdf"])
```
